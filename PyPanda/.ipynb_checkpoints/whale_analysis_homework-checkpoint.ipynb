```python
# Initial imports
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
import seaborn as sb

%matplotlib inline
```

```python
## Whale Returns

# Read the Whale Portfolio daily returns and clean the data.
whale_returns_path = Path("Resources/whale_returns.csv")
whale_returns_data = pd.read_csv(whale_returns_path, parse_dates=True,infer_datetime_format=True, index_col="Date")
whale_returns_data = whale_returns_data.sort_index()
whale_returns_data.dtypes


# Rename column name
whale_returns_data = whale_returns_data.rename(columns={
"SOROS FUND MANAGEMENT LLC": "Soros_Daily_Returns",
"PAULSON & CO.INC. ": "Paulson_Daily_Returns",
"TIGER GLOBAL MANAGEMENT LLC": "Tiger_Daily_Returns",
"BERKSHIRE HATHAWAY INC": "BerkshireHathaway_Daily_Returns",
})

# Cleanse data by dropping nulls
whale_returns_data = whale_returns_data.dropna().copy()

#determine number of nulls for whale_returns
whale_returns_data.isnull().sum()

```

```python
algo_returns_path = Path("Resources/algo_returns.csv")
algo_returns_data = pd.read_csv(algo_returns_path, parse_dates=True,infer_datetime_format=True, index_col="Date")

# sort by date
algo_returns_data = algo_returns_data.sort_index()

algo_returns_data.head()

# Rename column name
algo_returns_data = algo_returns_data.rename(columns={
    "Algo 1":"Algo1",
    "Algo 2":"Algo2"
})

# Cleanse data by dropping nulls
algo_returns_data = algo_returns_data.dropna().copy()

#determine number of nulls for whale_returns
algo_returns_data.isnull().sum()

algo_returns_data.head(200)
```

```python
sp_tsx_history_path = Path("Resources/sp_tsx_history.csv")
sp_tsx_history_data = pd.read_csv(sp_tsx_history_path, parse_dates=True,infer_datetime_format=True, index_col="Date")
sp_tsx_history_data = sp_tsx_history_data.sort_index()
sp_tsx_history_data.tail()

# Cleanse data by dropping nulls
sp_tsx_history_data = sp_tsx_history_data.dropna().copy()

#determine number of nulls for whale_returns
sp_tsx_history_data.isnull().sum()


# sp_tsx_history_data["Close"] = sp_tsx_history_data["Close"].astype("$","")
sp_tsx_history_data["Close"] = sp_tsx_history_data["Close"].str.replace("[^0-9]","",regex=True)
sp_tsx_history_data.head()
sp_tsx_history_data["Close"] = sp_tsx_history_data["Close"].astype("float")

# The whale portfolios and algorithmic portfolio CSV files contain daily returns, but the S&P TSX 60 CSV file contains closing prices. Convert the S&P TSX 60 closing prices to daily returns.
sp_tsx_history_daily = sp_tsx_history_data.pct_change()

sp_tsx_history_daily["Close"] = sp_tsx_history_daily["Close"].fillna(0)
sp_tsx_history_daily.head()
```

```python
## Join all 
# Join Whale Returns, Algorithmic Returns, and the S&P TSX 60 Returns into a single DataFrame with columns for each portfolio's returns.
joined_data_cols = pd.concat(
    [whale_returns_data, algo_returns_data, sp_tsx_history_daily], axis="columns", join="inner"
)
joined_data_cols.head(200)
```

```python
# Calculate and plot daily returns of all portfolios.
joined_data_cols.plot()
```

```python
# Calculate and plot cumulative returns for all portfolios. Does any portfolio outperform the S&P TSX 60?
cumulative_returns = (1 + joined_data_cols).cumprod() - 1
cumulative_returns.head()
cumulative_returns.plot()
```

```python
# Create a box plot for each of the returns.
# joined_data_cols.plot().box()
joined_data_cols.head()
```

```python
# Box plot to visually show risk
sb.set(rc={'figure.figsize':(10,5)})
bplot_daily_returns=sb.boxplot(data=joined_data_cols,width=0.5,palette="colorblind")
bplot_daily_returns.axes.set_title("Portfolio Daily Returns",fontsize=15)
bplot_daily_returns.set_xlabel("Portfolio", fontsize=13)
bplot_daily_returns.set_ylabel("Daily Returns",fontsize=13)
bplot_daily_returns.tick_params(labelsize=12)
bplot_daily_returns
# joined_data_cols.boxplot(column =['Soros_Daily_Returns','Paulson_Daily_Returns','Tiger_Daily_Returns','BerkshireHathaway_Daily_Returns'], grid = False)
```

```python
# Calculate the daily standard deviations of all portfolios and rank in order of volatility.Least to most volatile.
combined_daily_std_df = joined_data_cols.std()
combined_daily_std_df = combined_daily_std_df.sort_values(ascending=True)
combined_daily_std_df
```

```python
# Calculate the annualized standard deviation (252 trading days)
combined_annualized_std_df = combined_daily_std_df * np.sqrt(252)
combined_annualized_std_df

```

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
joined_data_cols.rolling(window=21).std()
```

```python
# Plot the rolling standard deviation
joined_data_cols.rolling(window=21).std().plot()
```

```python
# Calculate the correlation
combined_correlation_df = joined_data_cols.corr()
combined_correlation_df
# Display de correlation matrix
```
